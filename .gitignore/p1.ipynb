```python
import pandas as pd
import seaborn as sn
# import numpy as np
# import sklearn 
import matplotlib.pyplot as plt


dt = pd.read_csv("https://raw.githubusercontent.com/pyANSH/dsbda/main/dataset/prac1.csv")
```

```python
dt.head()
```

```python
dt.describe()
```

```python
dt.isna().sum()
```

```python
col=[]
miss = []
col.extend(dt.columns)
for i in col:
    t = dt[i].isnull().sum()
    if t !=0:
        miss.append(i)
    continue
```

```python
miss
```

```python
for j in miss:
    print(j)
    q = dt[j].dtypes
    if q == 'int64' or q == 'float64':
        dt[j].fillna(dt[j].mean(), inplace = True)
    else:
        
        dt[j].fillna(method="bfill", inplace=True)
```

```python
dt.isna().sum()
```

```python
plt.figure(figsize = (20,20))
sn.heatmap(dt.corr(), annot = True)
```

```python
# categorical variables into quantitative variables
gd = pd.get_dummies(dt, columns = ['horsepower-binned'])
gd
```
