```python
# p - 4
from sklearn.datasets import load_boston

boston = load_boston()

dt4 = pd.DataFrame(boston.data)
dt4.columns = boston.feature_names
```

```python
dt4['PRICE'] = boston.target
```

```python
dt4.head()
```

```python
sn.heatmap(dt4.corr(),annot = True)
```

```python
from sklearn.model_selection import train_test_split
x = boston.data
y = boston.target
```

```python
x_train, x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
pred = lr.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, pred)
# cm = confusion_matrix(y_test,pred)
```

```python
sn.distplot(y_test-pred)
```

```python
# p6
data = pd.read_csv('https://raw.githubusercontent.com/pyANSH/dsbda/main/dataset/iris_p6-p10.csv')
```
