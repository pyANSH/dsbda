```python
import pandas as pd 
import numpy as np
import seaborn as sn

dt = pd.read_csv("https://raw.githubusercontent.com/pyANSH/dsbda/main/dataset/covid_vaccine_statewise.csv")

dt.head()
```

```python
dt2 = dt
# dt2.isna().sum()
dt2.info()
```

```python
col = []
col.extend(dt2.columns)
```

```python
for i in col:
    q = dt[i].dtypes
    if (q == 'int64' or q == 'float64'):
        dt2[i].fillna(0, inplace=True)
    else:
        dt2[i].fillna(method = 'bfill', inplace=True)
```

```python
dt2.head()
```

```python
from matplotlib.pyplot import figure
figure(figsize=(20,20))
sn.barplot(x = 'First Dose Administered',y ='State',data=dt2)
```
