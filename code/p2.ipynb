```python
col = []
miss = []
col.extend(dt.columns)
for i in col:
    t = dt[i].isna().sum()
    if t != 0:
        miss.append(i)
    continue
    
col
```

```python
miss
```

```python
for j in miss:
    q = dt[j].dtypes
    if (q=='int64' or q == 'float64'):
        dt[j].fillna(dt[j].mean(), inplace=True)
    else:
        dt[j].fillna(method="ffill", inplace=True)
```

```python
dt['Total'] = dt['Phy_marks'] + dt['Che_marks'] + dt['EM1_marks'] + dt['PPS_marks'] +dt['SME_marks']
dt['percentage'] = dt['Total']/5
dt.head()
```
