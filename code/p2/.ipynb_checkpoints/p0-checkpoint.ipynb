```python
#  data wrangling can done through pandas
import pandas as pd 

```

```python
# import file

# df = pd.read_csv("./data.csv")
# df - to print entire file
# df.head() to print header only
# parse_dates it changes the date type of day
df = pd.read_csv("./data.csv", parse_dates=["day"])
df.head()
```

```python
type(df.day[0])

# returns types of values in a particular dataset
```

```python
# here index is 0,1,2,3,4
# to change this use set_index
df.set_index('day', inplace = True)
```

```python
df.head()
```

```python
# to replace na with some value
# new_df = df.fillna(0)  # this will replace the entire naN in file 
new_df = df.fillna({
    'temperature' : 0,
    'windspeed' : 0, 
    'event': 'no event'
}) # for specific 
new_df.head() 
```

```python
# to carry forword previous values
#     ffill = forword fill
#     bfill = backword fill
#     axis = change the direction of fill (row, column)
#     limit = limit the ffill to some ( like copy only one or 2 lines )

new_df = df.fillna(method = "ffill", limit=2)
new_df.head()
```

```python
# to insert average in missing values 
# we can see dates are different - to resolve that we will use method time 
new_df = df.interpolate(method="time")
new_df
```

```python
# drop all rows 
# new_df = df.dropna() #every cell with na will drop
new_df = df.dropna(thresh = 1) # if valid values >=  2  it won't drop it 
new_df
```

```python
# to insert missing rows for missing dates

dt = pd.date_range("01-01-2017" , "01-11-2017")
idt = pd.DatetimeIndex(dt) 
df = df.reindex(idt)
df
```

```python
new_df = df.fillna(method="ffill")
new_df
```

```python
new_df =  df.interpolate()
new_df
new_df.fillna('Sunny')
new_df
```
